Training and Testing Data

```python
X = admission.drop('Chance', axis=1)
y = admission['Chance']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lrm = LinearRegression()
```

```python
lrm.fit(X_train, y_train)
```

```python
lrm.coef_
```

```python
coefficients =pd.DataFrame(data=lrm.coef_, index=admission.drop('Chance', axis=1).columns)
coefficients = coefficients.rename(columns={0 : "Coeff"})
coefficients.style.background_gradient(cmap='Pastel1')
```

```python
pred_vals = lrm.predict(X_test)
```

```python
sns.scatterplot(x=y_test,y=pred_vals,hue=pred_vals, palette='Blues')
```

```python
from sklearn import metrics
```

```python
print('MAE: ',metrics.mean_absolute_error(y_test,pred_vals))
print('RSME: ',metrics.mean_squared_error(y_test,pred_vals))
```

```python
sns.histplot((y_test-pred_vals), stat='probability',kde=True)
```

```python
student_a = [332,120,4,4,5,9.7,1]
student_a = np.array(student_a).reshape(1,-120)
print("student_a has", lrm.predict(student_a) * 100, "percent chances of being picked")
```
